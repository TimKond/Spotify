# Add relevant json fields to dataframe

```python
# get all files 
from os import listdir
from os.path import isfile, join
path = "data/"
onlyfiles = [f for f in listdir(path) if isfile(join(path, f))]
```

```python
len(onlyfiles)
```

```python
import json

fields_to_extract = ["id", "name","popularity", "explicit","duration_ms",
                     "acousticness", "danceability","energy","instrumentalness","key","liveness","loudness","mode","speechiness","tempo","valence"]
deep_fields_to_extract = ["release_date", "prime_artist"]

extracted_fields = {}
for field in fields_to_extract + deep_fields_to_extract:
    extracted_fields = {**extracted_fields, **{field: []}}

for filename in onlyfiles:

    with open(path + filename, encoding='utf-8') as f:
        song = json.load(f)

    if not "error" in list(song.keys()): # some songs have -- status": 404, "message": "analysis not found"
        
        for field in fields_to_extract:
            extracted_fields[field].append(song[field])

        extracted_fields["release_date"].append(song["album"]["release_date"])

        extracted_fields["prime_artist"].append(song["artists"][0]["name"])
    
```

```python
import pandas as pd
  
# Create DataFrame
df = pd.DataFrame(extracted_fields)
  
# Print the output.
df.head()
```

```python
df.to_csv("spotify_song_features.csv", index=False)
```
